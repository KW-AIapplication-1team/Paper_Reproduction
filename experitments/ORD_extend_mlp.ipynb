```python
import os
import pandas as pd

# 경로 설정 (네 드라이브 구조에 맞게만 수정)
BASE_CTABSYN = "/content/drive/MyDrive/인공지능응용/ORD/ctabsyn"
BASE_ORD     = "/content/drive/MyDrive/인공지능응용/ORD"

syn0_path = os.path.join(BASE_CTABSYN, "data/adult_ORD_mlp_ord/tabsyn/syn_cond0.csv")
syn2_path = os.path.join(BASE_CTABSYN, "data/adult_ORD_mlp_ord/tabsyn/syn_cond2.csv")

syn0 = pd.read_csv(syn0_path)
syn2 = pd.read_csv(syn2_path)

print("syn0:", syn0.shape)
print("syn2:", syn2.shape)

# --- cond 컬럼 강제로 세팅 (없으면 새로 만들고, 있으면 값만 덮어씀) ---
if "cond" not in syn0.columns:
    syn0["cond"] = 0
else:
    syn0["cond"] = 0

if "cond" not in syn2.columns:
    syn2["cond"] = 2
else:
    syn2["cond"] = 2

# concat
syn_ord = pd.concat([syn0, syn2], ignore_index=True)

print("concat shape:", syn_ord.shape)
print("cond 분포:\n", syn_ord["cond"].value_counts())
```

```python
# ORD 쪽 원본(ternary) imbalanced 데이터
orig_ord_path = os.path.join(BASE_ORD, "data/adult_ORD_mlp/imbalanced_ord.csv")
orig_ord = pd.read_csv(orig_ord_path)

print("original imbalanced_ord shape:", orig_ord.shape)
print("original columns:", list(orig_ord.columns))

# 부족한 컬럼이 있으면 0으로 채워 넣기 (형태만 맞추기)
for col in orig_ord.columns:
    if col not in syn_ord.columns:
        # cond는 위에서 이미 채웠으니, 나머지는 0으로 기본값만 채워 둔다
        syn_ord[col] = 0

# 컬럼 순서를 imbalanced_ord와 동일하게 재정렬
syn_ord = syn_ord[orig_ord.columns]

print("aligned syn_ord shape:", syn_ord.shape)
print("aligned columns:", list(syn_ord.columns))
```

```python
save_dir_ord = os.path.join(BASE_ORD, "data/adult_ORD_mlp/tabsyn")
os.makedirs(save_dir_ord, exist_ok=True)

save_path_ord = os.path.join(save_dir_ord, "syn_ord.csv")
syn_ord.to_csv(save_path_ord, index=False)

print("저장 완료:", save_path_ord)
```
